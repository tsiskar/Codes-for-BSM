```python
perm = [0, 3, 1, 2]

n = 4
Pperm = sp.zeros(n)
# set Pperm[row, col] = 1 where col is original index, row is new index
for new_idx, old_idx in enumerate(perm):
    Pperm[new_idx, old_idx] = 1

sp.pprint(Pperm)   # permutation matrix

# apply permutation to get block form: H_blocked = Pperm * HB_4 * Pperm.T
# (This Pperm maps original->permuted basis; using this ordering we will diagonalize blocks)
H_blocked = Pperm * HB_4 * Pperm.T
sp.pprint(sp.simplify(H_blocked))

# Extract the two 2x2 blocks
B1 = sp.simplify(H_blocked[0:2, 0:2])
B2 = sp.simplify(H_blocked[2:4, 2:4])
print("\nBlock B1:")
sp.pprint(B1)
print("\nBlock B2:")
sp.pprint(B2)

# Build real orthogonal rotation for a 2x2 hermitian [[a, b],[b, c]]
def orthogonal_rotation_2x2(B):
    a = sp.simplify(B[0, 0])
    b = sp.simplify(B[0, 1])
    c = sp.simplify(B[1, 1])
    # use atan2 for robust quadrant handling: tan(2θ) = 2b/(a-c)
    two_theta = sp.atan2(2*b, a - c)           # two_theta = arctan2(2b, a-c)
    theta = sp.simplify(two_theta / 2)
    R = sp.Matrix([[sp.cos(theta), -sp.sin(theta)],
                   [sp.sin(theta),  sp.cos(theta)]])
    # Diagonal check
    D = sp.simplify(R.T * B * R)
    return R, sp.simplify(D), theta

R1, D1, th1 = orthogonal_rotation_2x2(B1)
R2, D2, th2 = orthogonal_rotation_2x2(B2)
R_f = R1@R2

print("\nRotation R1 (theta1):")
sp.pprint(sp.simplify(R1))
print("theta1 =", sp.simplify(th1))

print("\nRotation R2 (theta2):")
sp.pprint(sp.simplify(R2))
print("theta2 =", sp.simplify(th2))

print("\nD1 (should be diagonal):")
sp.pprint(D1)
print("\nD2 (should be diagonal):")
sp.pprint(D2)

# Build block-diagonal rotation in permuted basis
P_block = sp.BlockDiagMatrix(R1, R2).as_explicit()

# Total rotation in original basis:
# We used H_blocked = Pperm * HB_4 * Pperm.T, and P_block diagonalizes H_blocked:
# P_block.T * H_blocked * P_block = D_block
# Substitute H_blocked -> Pperm * HB_4 * Pperm.T:
# P_block.T * Pperm * HB_4 * Pperm.T * P_block = D_block
# => (Pperm.T * P_block).T * HB_4 * (Pperm.T * P_block) = D_block
# So final P_total that diagonalizes HB_4 is: P_total = Pperm.T * P_block
P_total = (Pperm.T * P_block).as_explicit()

print("\nP_total (single 4x4 rotation matrix):")
sp.pprint(sp.simplify(P_total))

# Verify diagonalization
D_full = sp.simplify(P_total.T * HB_4 * P_total)   # use transpose for real orthogonal
print("\nDiagonalized H (P_total.T * HB_4 * P_total):")
sp.pprint(sp.simplify(D_full))
```
